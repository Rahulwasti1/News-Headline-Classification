```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

Heading = []
Description = []

for i in range(1,100):

    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    
    r = requests.get("https://english.ratopati.com/category/lifestyle?page="+str(i), headers=headers)
    
    soup = BeautifulSoup(r.text, 'lxml')
    soup
    
    title = soup.find_all("h3", class_="post-card__title")
    title

    for i in title:
        text = i.text
        Heading.append(text.strip())

    desc = soup.find_all("div", class_="article-excerpt-default__teaser")

    for i in desc:
        text1 = i.text
        Description.append(text1.strip())

df = pd.DataFrame({"Heading":Heading, "Description":Description})

df.to_csv("news.csv")

```
